```python
genre_counts = df['playlist_genre'].value_counts()
fig1 = px.bar(x=genre_counts.index, y=genre_counts.values, color = genre_counts.index, color_discrete_sequence= px.colors.sequential.turbid,
              text = genre_counts.values,
              title = 'Distribution of songs among various genres')
fig1.update_layout(xaxis_title = 'Genre', yaxis_title = 'No of Songs')

fig1.show()
```

```python
popular_genre = df.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=popular_genre, x = 'track_popularity', y='playlist_genre', palette='cividis')
plt.title('Popularity of different genres')
plt.xlabel('Track Popularity')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize = (8,4))
plt.subplot(1,2,1)
plt.style.use('seaborn-v0_8-pastel')
decade_50s = df.loc[df.release_year < 1960].count()[0]
decade_60s = df.loc[(df.release_year >= 1960) & (df.release_year < 1970)].count()[0]
decade_70s = df.loc[(df.release_year >= 1970) & (df.release_year < 1980)].count()[0]
decade_80s = df.loc[(df.release_year >= 1980) & (df.release_year < 1990)].count()[0]
decade_90s = df.loc[(df.release_year >= 1990) & (df.release_year < 2000)].count()[0]
decade_2k = df.loc[(df.release_year >= 2000) & (df.release_year < 2010)].count()[0]
decade_2010s = df.loc[(df.release_year >= 2010) & (df.release_year < 2020)].count()[0]
decade_2020s = df.loc[(df.release_year >= 2020)].count()[0]
decades = [decade_70s,decade_80s,decade_90s,decade_2k,decade_2010s]
labels = ['1970s','1980s','1990s','2000s','2010s']
explode = (0,0,0,0,0.2)
plt.title('Distribution of Songs Released Each Decade',fontsize = 10,fontweight = 'bold')
plt.pie(decades,labels = labels,autopct = '%1.1f %%',pctdistance = 0.8,
        explode = explode,startangle = 305,colors = plt.cm.Paired.colors,textprops = {'fontsize':7})
plt.subplot(1,2,2)
xticks = [1950,1960,1970,1980,1990,2000,2010,2020]
ydecades = [decade_50s,decade_60s,decade_70s,decade_80s,decade_90s,decade_2k,decade_2010s,decade_2020s]
plt.plot(xticks,ydecades,marker = 'o',linestyle = '-',color = '#6abaa2',label = 'Songs Released')
plt.title('Songs Released Each Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Songs')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python


from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2['playlist_genre'] = le.fit_transform(df2['playlist_genre'])
df2['playlist_subgenre'] = le.fit_transform(df2['playlist_subgenre'])
df2['track_artist'] = le.fit_transform(df2['track_artist'])
df2['track_name'] = le.fit_transform(df2['track_name'])
```

```python
df_pr = df2.drop(['track_popularity','track_name','track_album_name','track_id','playlist_id','track_album_release_date','track_album_id','playlist_name',\
    'track_artist','loudness','playlist_subgenre','duration_ms','year','instrumentalness','mode','key'], axis=1)
```

```python
x = df_pr.drop('valence', axis=1)
y = df_pr.valence
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score, accuracy_score

x = df_pr.drop('valence', axis=1)
y = df_pr.valence

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model1 = RandomForestRegressor()
model2 = KNeighborsRegressor()
model3 = LinearRegression()
model4 = DecisionTreeRegressor()
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
```

```python
print(f'Random Forest: {model1.score(X_test, y_test)}')
print(f'KNN: {model2.score(X_test, y_test)}')
print(f'Linear Regression: {model3.score(X_test, y_test)}')
print(f'Decision Tree: {model4.score(X_test, y_test)}')
print(f'Random Forest R-squared: {r2_score(y_test, model1.predict(X_test))}')
print(f'Random Forest MSE: {mean_squared_error(y_test, model1.predict(X_test))}')
print(f'KNN R-squared: {r2_score(y_test, model2.predict(X_test))}')
print(f'KNN MSE: {mean_squared_error(y_test, model2.predict(X_test))}')
print(f'Linear Regression R-squared: {r2_score(y_test, model3.predict(X_test))}')
print(f'Linear Regression MSE: {mean_squared_error(y_test, model3.predict(X_test))}')
print(f'Decision Tree R-squared: {r2_score(y_test, model4.predict(X_test))}')
print(f'Decision Tree MSE: {mean_squared_error(y_test, model4.predict(X_test))}')
```
